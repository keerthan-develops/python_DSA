```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1

    def append(self, value):
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node
        self.length += 1
        return True
    
    def prepend(self, value):
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            self.head.prev = new_node
            new_node.next = self.head
            self.head = new_node
        self.length += 1
        return True
    
    def insert(self, index, value):        
        if index > self.length or index < -1:
            return False
        elif index == 0:
            self.prepend(value)
            return True
        elif index == self.length:
            self.append(value)
            return True
        elif index > self.length//2:
            new_node = Node(value)
            temp_node = self.tail
            for i in range(self.length, index + 1):
                temp_node = temp_node.prev
            next_node = temp_node
            prev_node = temp_node.prev
            prev_node.next = new_node
            next_node.prev = new_node
            new_node.next = next_node
            new_node.prev = prev_node
            self.length += 1
            return True
        else:
            new_node = Node(value)
            temp_node = self.head
            for i in range(0, index - 1):
                temp_node = temp_node.next
            prev_node = temp_node
            next_node = temp_node.next
            prev_node.next = new_node
            next_node.prev = new_node
            new_node.next = next_node
            new_node.prev = prev_node
            self.length += 1
            return True
    
    def print_DoublyLinkedList(self):
        current_node = self.head
        while current_node is not None:
            print(current_node.value)
            current_node = current_node.next
        return True
        
dll = DoublyLinkedList(30)

print('-> Append')
dll.append(40)

print('-> prepend')
dll.prepend(20)
dll.prepend(10)
dll.append(50)
dll.append(60)
dll.insert(6, 70)
dll.insert(0, 1)
dll.insert(2, 14)
dll.insert(7, 64)

dll.print_DoublyLinkedList()
```
