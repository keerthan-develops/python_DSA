```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0
    
    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1

    def __str__(self):
        temp_node = self.head
        result = ''
        while temp_node is not None:
            result += str(temp_node.value)
            #print('Start :')
            #print(temp_node)
            #print(temp_node.value)
            #print(temp_node.next)
            if temp_node.next is not None:
                result += ' -> '
            temp_node = temp_node.next
        #print(temp_node)    
        return result

    def append(self, value):                           # 0(1)
        new_node = Node(value)          # 0(1)
        if self.head is None:           # 0(1)
            self.head = new_node        # 0(1)
            self.tail = new_node        # 0(1)
        else:
            self.tail.next = new_node   # 0(1)
            self.tail = new_node        # 0(1)
        self.length += 1                # 0(1)
        return True

    def prepend(self, value):  # 0(1)
        new_node = Node(value)          # 0(1)
        if self.head is None:           # 0(1)
            self.head = new_node        # 0(1)
            self.tail = new_node        # 0(1)
        else:
            new_node.next = self.head   # 0(1)
            self.head = new_node        # 0(1)
        self.length += 1                # 0(1)
        return True

    def insert(self, index, value):   # 0(n)
        new_node = Node(value)                      # 0(1)
        if index < 0 or index > self.length:
            return False
        elif self.head is None:
            self.head = new_node
            self.tail = new_node
        elif index == 0:
            new_node.next = self.head
            self.head = new_node
        elif index == self.length:
            self.tail.next = new_node
            self.tail = new_node
        else:
            temp_node = self.head               # 0(1)
            for _ in range(0, index-1):                 # 0(n)
                temp_node = temp_node.next      # 0(1)
            new_node.next = temp_node.next      # 0(1)
            temp_node.next = new_node           # 0(1)
        self.length += 1                        # 0(1)
        return True
    
    def traverse(self):                                  # 0(n)
        current = self.head                     # 0(1)
        while current is not None:                      # 0(n)
            print(current.value)                # 0(1)
            current = current.next              # 0(1)
        return True
    
    def search(self, target):   # 0(n)
        current = self.head
        index = 0
        while current is not None:                      # 0(n)            
            if current.value == target:         # 0(1)
                return index                    # 0(1)
            current = current.next              # 0(1)
            index += 1                          # 0(1)
        return False
    
    def get(self, index):        
        if index == -1:
            return self.tail
        if index > self.length or index < -1:
            return False
        else:
            current = self.head
            for _ in range(index):
                current = current.next
            return current
    
    def set_value(self, index, value):
        temp = self.get(index)
        if temp:
            temp.value = value
            return True    
        return False
    
    def pop_first(self):
        if self.length == 0:
            return None
        popped_node = self.head
        if self.length == 1:
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
            popped_node.next = None
        self.length -= 1
        return popped_node
    
    def pop(self):
        if self.length == 0:
            return None
        popped_node = self.tail
        if self.length == 1:
            self.head = self.tail = None
        else:
            temp = self.head
            while temp.next is not self.tail:
                temp = temp.next
            self.tail = temp
            temp.next = None
        self.length -= 1
        return popped_node
    
    def remove(self, index):
        if self.length == 0 or index >= self.length or index < -1:
            return None
        if self.length == 1:
            self.head = self.head = None
        if index == 0:
            return self.pop_first()
        if index == self.length - 1 or index == -1:
            return self.pop()
        
        prev_node = self.get(index-1)
        popped_node = prev_node.next
        prev_node.next = popped_node.next
        popped_node.next = None
        return popped_node
            
```

```python
ll = LinkedList(3)
#print(f'Head ->  {ll.head.value}')
#print(f'Tail ->  {ll.tail.value}')

print('-> Append')
ll.append(20)
#print(f'Head ->  {ll.head.value}')
#print(f'Tail ->  {ll.tail.value}')

print('-> prepend')
ll.prepend(4)
#print(f'Head ->  {ll.head.value}')
#print(f'Tail ->  {ll.tail.value}')

print('-> Insert middle')
ll.insert(2,6)
#print(f'Head ->  {ll.head.value}')
#print(f'Tail ->  {ll.tail.value}')

print('-> Insert beginning')
ll.insert(0,10)
#print(f'Head ->  {ll.head.value}')
#print(f'Tail ->  {ll.tail.value}')

print('-> Traverse')
ll.traverse()

print('-> Insert end')
ll.insert(5,12)
#print(f'Head ->  {ll.head.value}')
#print(f'Tail ->  {ll.tail.value}')

print('-> Traverse')
ll.traverse()

print('-> Search')
print(ll.search(100))


print('-> Get')
print(ll.get(-1))


print('-> Set_value')
print(ll.set_value(4, 1202))
print('-> Traverse')
ll.traverse()

print('-> pop first')
print(ll.pop_first())
print('-> Traverse')
ll.traverse()

print('-> Pop')
print(ll.pop())

print('-> Pop')
print(ll.pop())

print('Print LinkedList')
print(ll)

#ll.print_list()
```
